```python
import csv
import requests
import json
```

```python
column_names = []
cats_objectids = []
cats_objectids_int = []

with open('MetObjects.csv', 'r') as data:
    processed_csv = csv.DictReader(data, delimiter= ',')

    for row in processed_csv:
        column_names.append(row)
        break

    for row in processed_csv:
        if 'Cats' in row['Tags']:
            cats_objectids.append(row['Object ID'])

for item in cats_objectids:
    cats_objectids_int.append(int(item))

print(cats_objectids_int)
print(len(cats_objectids_int))
```

above, i created a list of all Object IDs where 'cats' is included in the tags

```python
cat_art_urls = []

for id in cats_objectids_int:
    base_url = "https://collectionapi.metmuseum.org/public/collection/v1/objects/" + str(id)
    cat_art_urls.append(base_url)

print(cat_art_urls)
print(len(cat_art_urls))
```

here, i used a loop to create a list of all urls i need to use to retrieve object data using the api

```python
cat_art_data = []

for url in cat_art_urls:
    r = requests.get(url)
    data = json.loads(r.text)
    if len(data['primaryImage']) != 0:
        cat_art_data.append(data)

print(cat_art_data)
```

```python
for obj in cat_art_data:
    obj.pop('accessionNumber')
    obj.pop('accessionYear')
    obj.pop('isHighlight')
    obj.pop('isPublicDomain')
    obj.pop('primaryImageSmall')
    obj.pop('additionalImages')
    obj.pop('constituents')
    obj.pop('department')
    obj.pop('objectName')
    obj.pop('period')
    obj.pop('dynasty')
    obj.pop('reign')
    obj.pop('portfolio')
    obj.pop('artistRole')
    obj.pop('artistPrefix')
    obj.pop('artistDisplayBio')
    obj.pop('artistSuffix')
    obj.pop('artistAlphaSort')
    obj.pop('artistNationality')
    obj.pop('artistBeginDate')
    obj.pop('artistEndDate')
    obj.pop('artistGender')
    obj.pop('artistWikidata_URL')
    obj.pop('artistULAN_URL')
    obj.pop('objectBeginDate')
    obj.pop('objectEndDate')
    obj.pop('dimensions')
    obj.pop('measurements')
    obj.pop('creditLine')
    obj.pop('geographyType')
    obj.pop('city')
    obj.pop('state')
    obj.pop('county')
    obj.pop('country')
    obj.pop('region')
    obj.pop('subregion')
    obj.pop('locale')
    obj.pop('locus')
    obj.pop('excavation')
    obj.pop('river')
    obj.pop('classification')
    obj.pop('rightsAndReproduction')
    obj.pop('linkResource')
    obj.pop('metadataDate')
    obj.pop('repository')
    obj.pop('objectURL')
    obj.pop('tags')
    obj.pop('objectWikidata_URL')
    obj.pop('isTimelineWork')
    obj.pop('GalleryNumber')

print(cat_art_data)
print(len(cat_art_data))
```

above, i used .pop() function to remove any data i did not want to display in my random generator. this function removes keys from a dictionary

```python
with open('MetCatsObjects.json', 'w') as out:
    json.dump(cat_art_data,out,indent=2)
```
